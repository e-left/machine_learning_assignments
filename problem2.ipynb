**4** Αναπτύξτε κώδικα με βάση την ιδιότητα [tree_](https://scikit-learn.org/stable/auto_examples/tree/plot_unveil_tree_structure.html#sphx-glr-auto-examples-tree-plot-unveil-tree-structure-py) του παρακάτω μοντέλου, προκειμένου να εμφανίσετε το μονοπάτι που ακολουθήθηκε στο δένδρο για την πρόβλεψη της πρώτης περίπτωσης του συνόλου ελέγχου. Σε περίπτωση που ελέγχεται δεύτερη φορά μια μεταβλητή στο μονοπάτι, θα πρέπει να διατηρούμε τον τελευταίο και στενότερο περιορισμό. Ακολουθούν παραδείγματα ορθής λειτουργίας, στα οποία φαίνεται και η μορφή αν-τότε που πρέπει να έχει το μονοπάτι. (4 μονάδες)

```python
from sklearn.tree import export_text
model = DecisionTreeRegressor(max_depth=3, random_state=42)
model.fit(X_train, y_train)
```

```python
def get_path(tree, instance, feature_names):
    # If passing one instance, add a dimension for correct shapes
    if len(instance.shape) == 1:
        instance = instance.reshape(1, -1)
    
    # Obtain threshold and feature values
    threshold = tree.tree_.threshold
    feature = tree.tree_.feature

    # Obtain the node indices for the decision path
    node_indicator = tree.decision_path(instance)
    leaf_id = tree.apply(instance)
    node_index = node_indicator.indices[
        node_indicator.indptr[0] : node_indicator.indptr[1]
    ]

    # Create a list to hold all rules
    rules = []

    # Iterate over all nodes
    for node_id in node_index:
        # continue to the next node if it is a leaf node
        if leaf_id[0] == node_id:
            continue

        # Obtain feature name
        feature_name = feature_names[feature[node_id]]
        # Obtain feature threshold
        feature_threshold = threshold[node_id]
        # Obtain feature value
        feature_value = instance[0, feature[node_id]]

        # check if value of the split feature for sample 0 is below threshold
        if feature_value <= feature_threshold:
            threshold_sign = "<="
        else:
            threshold_sign = ">"
        
        # Iterate over rules list, finding if this rule is before, further constricting it if needed
        found_rule = False
        for rule_idx in range(len(rules)):
            # Get the current rule
            rule = rules[rule_idx]
            # If a check is found
            if rule[0] == feature_name and rule[1] == threshold_sign:
                # To ensure addition does not happen later
                found_rule = True
                # Update the threshold, performing the correct check
                # This is redundand in the decision tree, but added for clarity
                if threshold_sign == "<=" and feature_threshold <= rule[2]:
                    rules[rule_idx][2] = feature_threshold
                if threshold_sign == ">" and feature_threshold > rule[2]:
                    rules[rule_idx][2] = feature_threshold
            
        if not found_rule:
            rules.append([feature_name, threshold_sign, feature_threshold])

    
    # Construct result text
    result_text = ""

    # Loop over all found rules
    for rule in rules:
        if result_text == "":
            starting_text = "αν "
        else:
            starting_text = "και "
        
        string_segment = starting_text + rule[0] + " " + rule[1] + " " + str(round(rule[2], 2)) + " "
        result_text += string_segment

    # Add final value
    result_text += "τότε " + str(round(tree.predict(instance)[0], 2))
    return(result_text)
```

```python
"""Τεστ ορθού υπολογισμού του μονοπατιού"""
assert get_path(model, X_test[0], diabetes.feature_names) == 'αν bmi <= 0.01 και s5 > 0.01 και s4 <= 0.09 τότε 159.57'
assert get_path(model, X_test[1], diabetes.feature_names) == 'αν bmi > 0.01 και bmi <= 0.07 και s6 <= 0.03 τότε 175.8'
assert get_path(model, X_test[6], diabetes.feature_names) == 'αν bmi > 0.07 και s2 > 0.02 τότε 225.75'
```

**6** Υλοποιήστε συνάρτηση, η οποία δοθέντος ενός δενδρικού μοντέλου παλινδρόμησης, διερευνά όλες τις ακέραιες τιμές της μεταβλητής min_samples_leaf από ένα κάτω όριο μέχρι ένα άνω όριο και επιστρέφει την βέλτιστη με βάση το mean squared error στο παραπάνω σύνολο επικύρωσης έπειτα από εκπαίδευση στο παραπάνω σύνολο εκπαίδευσης. Επιπλέον επιστρέφει και το mean squared error (στρογγυλοποιημένο σε ακέραιο) που πετυχαίνει στο σύνολο ελέγχου το μοντέλο με την βέλτιστη επίδοση στο σύνολο επικύρωσης. (2 μονάδες)

```python
from sklearn.metrics import mean_squared_error

def best_min_samples_leaf(model, min, max):
    # Initialize a variable to hold the best min samples leaf, as well as the best error found both in validation and test sets
    best_msl = 0
    best_val_mse = float('inf')
    best_test_mse = float('inf')
    
    # Iterate over all possible min sample leaf values
    for msl in range(min, max + 1):
        # Get a copy of the model, with the parameter min sample leaf set
        this_model = model.set_params(min_samples_leaf=msl)
        # Fit the model to the training data, and make predictions on the validation set
        this_model.fit(X_train, y_train)
        y_pred = this_model.predict(X_val)
        # Calculate MSE metrics on test and validation set
        val_mse = mean_squared_error(y_val, y_pred)
        test_y_pred = this_model.predict(X_test)
        test_mse = mean_squared_error(y_test, test_y_pred)
        
        # Keep the metrics for the best validation MSE
        if val_mse < best_val_mse:
            best_val_mse = val_mse
            best_msl = msl
            best_test_mse = test_mse
    
    # Return the best min sample leaf value, along with the MSE on the test set for that value
    return (best_msl, round(best_test_mse))
```

```python
"""Τεστ ορθής υλοποίησης συνάρτησης"""
model = DecisionTreeRegressor(max_depth=3, random_state=42)
assert best_min_samples_leaf(model, 1, 8) == (7, 3358)
assert best_min_samples_leaf(model, 5, 15) == (11, 3271)
```

**7** Υλοποιήστε συνάρτηση η οποία δοθέντος ενός μοντέλου παλινδρόμησης, ενός συνόλου δεδομένων και μίας μετρικής, το αξιολογεί με τη μέθοδο της σταυρωτής επικύρωσης "άφησε ένα εκτός". (1 μονάδα)

```python
from sklearn.model_selection import LeaveOneOut
import numpy as np

def leave_one_out(model, X, y, metric):
    # Initialize leave one out splitter
    loo = LeaveOneOut()
    scores = []
    
    # Iterate over all folds
    for train_idx, val_idx in loo.split(X):
        # Use separate model each time to ensure tests are independent
        this_model = model
        X_train, X_val = X[train_idx], X[val_idx]
        y_train, y_val = y[train_idx], y[val_idx]
        
        # Fit the model to the training data, and make predictions on the test set
        this_model.fit(X_train, y_train)
        y_pred = this_model.predict(X_val)
        
        # Calculate score via the provided metric function
        score = metric(y_val, y_pred)
        
        # Add this score to the list
        scores.append(score)
    
    # Return mean of all scores, evaluating the regressor
    return np.mean(scores)
```
